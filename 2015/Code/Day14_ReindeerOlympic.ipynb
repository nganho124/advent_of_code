```python
import pandas as pd
import re
```

```python
input = pd.read_excel("../DATA/Day14_ReindeerOlympic.xlsx")
```

```python
input[['time','rest']] = input['reindeers'].str.split(', ', expand=True)
input['name'] = input['time'].str.split(' ').str[0]
# input[['speed', 'f_duration']] = re.findall(r'\d+', input['time'])
# input[['speed', 'f_duration']] = 
extract = input['time'].str.extractall(r'(\d+)')
```

```python
speed = []
f_duration = []
for i in range(9):

    speed.append(extract[0][i][0])
    f_duration.append(extract[0][i][1])

input['speed'] = speed
input['f_duration'] = f_duration
input['r_duration'] = input['rest'].str.extract(r'(\d+)')
```

```python
input[['speed', 'f_duration', 'r_duration']] = input[['speed', 'f_duration', 'r_duration']].astype(int)
```

# Part 1

```python
time_period = 2503
distance_dict = {
    'name': [],
    'distance': []
}
```

```python
1000 // 137
```

```python
for index, row in input.iterrows():

    distance_dict['name'].append(row['name'])

    total_time = row['f_duration'] + row['r_duration']
    n_seq = time_period - (time_period // total_time)*total_time

    if n_seq >= row['f_duration']:

        distance = (time_period // total_time + 1)*row['f_duration']*row['speed']
        distance_dict['distance'].append(distance)
        
    else:

        distance = (time_period // total_time)*row['f_duration']*row['speed'] + n_seq*row['speed']
        distance_dict['distance'].append(distance)
```

# Part 2

```python
time_period = 2503
point_dict = {}
for i in range(len(input)):

    point_dict[input['name'][i]] = 0
        
```

```python
# s = 1
```

```python
for s in range(1, time_period + 1):

    check_1st = {
        'name': [],
        'distance': []
    }
    for index, row in input.iterrows():

        check_1st['name'].append(row['name'])

        total_time = row['f_duration'] + row['r_duration']
        n_seq = s - (s // total_time)*total_time

        if n_seq >= row['f_duration']:

            distance = (s // total_time + 1)*row['f_duration']*row['speed']
            check_1st['distance'].append(distance)
            
        else:

            distance = (s // total_time)*row['f_duration']*row['speed'] + n_seq*row['speed']
            check_1st['distance'].append(distance)

    check_1st_df = pd.DataFrame(check_1st)
    first = check_1st_df[check_1st_df.distance == check_1st_df.distance.max()]

    for name in first['name']:

        point_dict[name] += 1
```
