```python
import pandas as pd
```

```python
lobby = pd.read_csv("../DATA/Day03_Lobby.csv", names = ["Lobby"])
```

## Part 1

```python
sum = 0
list_max = []
for i, row in lobby.iterrows():
    max = int(row['Lobby'][0] + row['Lobby'][1])

    for j in range(len(row['Lobby']) - 1):

        for z in range(j + 1, len(row['Lobby'])):
            
            if int(row['Lobby'][j] + row['Lobby'][z]) > max:
                max = int(row['Lobby'][j] + row['Lobby'][z])

    
    list_max.append(max)    
    sum += max
```
